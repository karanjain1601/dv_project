```python
for i in range(len(data)):
    idx = data["ID"][i]
    team1_val = data2.loc[data2["ID"]==idx, "Team1"].values
    team2_val = data2.loc[data2["ID"]==idx, "Team2"].values
    
    if data.loc[i, "BattingTeam"] is None or data.loc[i, "BattingTeam"] == team1_val:
        data.loc[i, "BowlingTeam"] = team2_val[0]
    else:
        data.loc[i, "BowlingTeam"] = team1_val[0]
```
